```python
import numpy as np
import matplotlib.pyplot as plt
import pickle

# 加载数据
with open('../../data/main_figures/slr_res/DIVA_data/Coast_with_Mang.pkl', 'rb') as f:
    data = pickle.load(f)

# 选择Pop5和恢复策略50%
area_after_SLR_126_1 = data['area_after_SLR_126'][:, 1]
area_after_SLR_585_1 = data['area_after_SLR_585'][:, 1]
area_restoration_126_1 = data['area_restoration_126'][:, 0]
area_restoration_585_1 = data['area_restoration_585'][:, 0]

# 提取agb和area数据
agb_present = data['agb_present']
agb_grow = data['agb_grow']
agb_126 = data['agb_126']
agb_585 = data['agb_585']
area = data['area']

# 计算各类总AGB
total_agb_present = np.nansum(agb_present * area) / (10 ** 9)
total_agb_grow = np.nansum(agb_grow * area) / (10 ** 9)
total_agb_126 = np.nansum(agb_126 * area) / (10 ** 9)
total_agb_585 = np.nansum(agb_585 * area) / (10 ** 9)

total_agb_126_restoration = np.nansum(agb_126 * (area + area_restoration_126_1)) / (10 ** 9)
total_agb_585_restoration = np.nansum(agb_585 * (area + area_restoration_585_1)) / (10 ** 9)
total_agb_126_restoration_SLR = np.nansum(agb_126 * (area_after_SLR_126_1 + area_restoration_126_1)) / (10 ** 9)
total_agb_585_restoration_SLR = np.nansum(agb_585 * (area_after_SLR_585_1 + area_restoration_585_1)) / (10 ** 9)

# Set global font to Helvetica
plt.rcParams['font.family'] = 'Helvetica'
plt.rcParams['font.size'] = 16
# Set the mathtext font to Helvetica
plt.rcParams['mathtext.fontset'] = 'custom'
plt.rcParams['mathtext.rm'] = 'Helvetica'
plt.rcParams['mathtext.it'] = 'Helvetica:italic'
plt.rcParams['mathtext.bf'] = 'Helvetica:bold'
```

```python
# 绘制图形
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
fig.subplots_adjust(hspace=0.4, wspace=0.4)

# 第一张图：低温化情境
xtips = np.arange(1, 7)
ytips = [(total_agb_present + 1) / 2, total_agb_grow, total_agb_126, total_agb_126_restoration, total_agb_126_restoration_SLR, (total_agb_126_restoration_SLR + 1) / 2]
labels = [f"{total_agb_present:.2f}",
          f"{total_agb_grow - total_agb_present:+.2f}",
          f"{total_agb_126 - total_agb_grow:+.2f}",
          f"{total_agb_126_restoration - total_agb_126:+.2f}",
          f"{total_agb_126_restoration_SLR-total_agb_126_restoration:+.2f}",
          f"{total_agb_126_restoration_SLR:.2f}"]

# 核心修改：使用 bottom 参数，取消所有的白色遮挡柱
bars1 = [
    ax1.bar(1, total_agb_present, color=(204 / 255, 204 / 255, 204 / 255), edgecolor='none'),
    ax1.bar(2, total_agb_grow - total_agb_present, bottom=total_agb_present, color=(118 / 255, 215 / 255, 196 / 255), edgecolor='none', label='Potential growth to maturity'),
    ax1.bar(3, total_agb_126 - total_agb_grow, bottom=total_agb_grow, color=(243 / 255, 182 / 255, 177 / 255), edgecolor='none', label='Climate and hydrology'),
    ax1.bar(4, total_agb_126_restoration - total_agb_126, bottom=total_agb_126, color=(235 / 255, 222 / 255, 247 / 255), edgecolor='none', label='Restoration'),
    ax1.bar(5, total_agb_126_restoration_SLR - total_agb_126_restoration, bottom=total_agb_126_restoration, color=(174 / 255, 214 / 255, 241 / 255), edgecolor='none', label='Sea-level rise'),
    ax1.bar(6, total_agb_126_restoration_SLR, color=(204 / 255, 204 / 255, 204 / 255), edgecolor='none')
]

# 设置每个矩形对象的边框颜色
for bar in bars1:
    for rect in bar.get_children():
        rect.set_edgecolor('none')

# 添加文本标签
for i, (xtip, ytip, label) in enumerate(zip(xtips, ytips, labels)):
    label_value = float(label)
    if label_value > 0:
        ax1.text(xtip, ytip, label, ha='center', va='bottom')
    else:
        ax1.text(xtip, ytip - 0.01, label, ha='center', va='top')

# 第二张图：高温化情境
xtips = np.arange(1, 7)
ytips = [(total_agb_present + 1) / 2, total_agb_grow, total_agb_585, total_agb_585_restoration, total_agb_585_restoration_SLR, (total_agb_585_restoration_SLR + 1) / 2]
labels = [f"{total_agb_present:.2f}",
          f"{total_agb_grow - total_agb_present:+.2f}",
          f"{total_agb_585 - total_agb_grow:+.2f}",
          f"{total_agb_585_restoration - total_agb_585:+.2f}",
          f"{total_agb_585_restoration_SLR-total_agb_585_restoration:+.2f}",
          f"{total_agb_585_restoration_SLR:.2f}"]

# 核心修改：同上，使用 bottom 参数
bars2 = [
    ax2.bar(1, total_agb_present, color=(204 / 255, 204 / 255, 204 / 255), edgecolor='none'),
    ax2.bar(2, total_agb_grow - total_agb_present, bottom=total_agb_present, color=(118 / 255, 215 / 255, 196 / 255), edgecolor='none', label='Growth'),
    ax2.bar(3, total_agb_585 - total_agb_grow, bottom=total_agb_grow, color=(243 / 255, 182 / 255, 177 / 255), edgecolor='none', label='Climate and hydrology'),
    ax2.bar(4, total_agb_585_restoration - total_agb_585, bottom=total_agb_585, color=(235 / 255, 222 / 255, 247 / 255), edgecolor='none', label='Restoration'),
    ax2.bar(5, total_agb_585_restoration_SLR - total_agb_585_restoration, bottom=total_agb_585_restoration, color=(174 / 255, 214 / 255, 241 / 255), edgecolor='none', label='Sea-level rise'),
    ax2.bar(6, total_agb_585_restoration_SLR, color=(204 / 255, 204 / 255, 204 / 255), edgecolor='none')
]

# 设置每个矩形对象的边框颜色
for bar in bars2:
    for rect in bar.get_children():
        rect.set_edgecolor('none')

# 添加文本标签
for i, (xtip, ytip, label) in enumerate(zip(xtips, ytips, labels)):
    label_value = float(label)
    if label_value > 0:
        ax2.text(xtip, ytip, label, ha='center', va='bottom')
    else:
        ax2.text(xtip, ytip - 0.01, label, ha='center', va='top')

# 设置图形属性
for ax in [ax1, ax2]:
    ax.set_xlim([0, 7])
    ax.set_ylim([1.0, 2.5])
    ax.set_yticks(np.arange(1.0, 2.6, 0.5))
    ax.set_xticks([1, 6])
    ax.set_xticklabels(['Present-day', '2100'])
    ax.tick_params(axis='both')
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['bottom'].set_visible(False)

ax1.set_ylabel('Total AGB (Pg DM)')

# 添加图例
ax1.text(0.4, 2.45, 'Low-warming')
ax1.set_title('a', loc='left', fontweight='bold')
ax2.text(0.4, 2.45, 'High-warming')
ax2.set_title('b', loc='left', fontweight='bold')

# 获取所有图例句柄和标签
handles, labels = ax1.get_legend_handles_labels()
ax1.legend(handles, labels, loc='lower center', frameon=False, handlelength=0.8, fontsize=16, handletextpad=0.3)

# 保存图像
plt.tight_layout()
plt.savefig('py02_future_change_1.png', dpi=400, transparent=True)
plt.show()
```

```python
# --- Low-warming scenario (SSP1-2.6) ---
print("--- Low-warming scenario (SSP1-2.6) ---")
change_descriptions_126 = [
    "Present-day AGB",
    "Change from potential growth",
    "Change from hydro-thermal effects",
    "Change from restoration",
    "Change from sea-level rise"
]
values_126 = [
    total_agb_present,
    total_agb_grow - total_agb_present,
    total_agb_126 - total_agb_grow,
    total_agb_126_restoration - total_agb_126,
    total_agb_126_restoration_SLR - total_agb_126_restoration
]

# Print changes with aligned formatting
for desc, val in zip(change_descriptions_126, values_126):
    proportion = val / total_agb_present
    print(f"{desc:<35}: {val:+.2f} Pg DM, Proportion: {proportion:+.2%}")

# Calculate and print net change
net_change_126 = total_agb_126_restoration_SLR - total_agb_present
net_proportion_126 = net_change_126 / total_agb_present
print(f"{'Final AGB in 2100':<35}: {total_agb_126_restoration_SLR:.2f} Pg DM")
print(f"{'Net change by 2100':<35}: {net_change_126:+.2f} Pg DM, Proportion of change: {net_proportion_126:+.2%}")


# --- High-warming scenario (SSP5-8.5) ---
print("\n--- High-warming scenario (SSP5-8.5) ---")
change_descriptions_585 = [
    "Present-day AGB",
    "Change from potential growth",
    "Change from hydro-thermal effects",
    "Change from restoration",
    "Change from sea-level rise"
]
values_585 = [
    total_agb_present,
    total_agb_grow - total_agb_present,
    total_agb_585 - total_agb_grow,
    total_agb_585_restoration - total_agb_585,
    total_agb_585_restoration_SLR - total_agb_585_restoration
]

# Print changes with aligned formatting
for desc, val in zip(change_descriptions_585, values_585):
    proportion = val / total_agb_present
    print(f"{desc:<35}: {val:+.2f} Pg DM, Proportion: {proportion:+.2%}")

# Calculate and print net change
net_change_585 = total_agb_585_restoration_SLR - total_agb_present
net_proportion_585 = net_change_585 / total_agb_present
print(f"{'Final AGB in 2100':<35}: {total_agb_585_restoration_SLR:.2f} Pg DM")
print(f"{'Net change by 2100':<35}: {net_change_585:+.2f} Pg DM, Proportion of change: {net_proportion_585:+.2%}")
```

```python
# 低升温情境下恢复增加的AGB
agb_increase_restoration_126 = total_agb_126_restoration - total_agb_126

# 高升温情境下恢复增加的AGB
agb_increase_restoration_585 = total_agb_585_restoration - total_agb_585

# 计算高升温情境比低升温情境少的AGB量
difference = agb_increase_restoration_126 - agb_increase_restoration_585

# 计算减少的比例
percentage_less = (difference / agb_increase_restoration_126) * 100

print(f"低升温情境下恢复增加的AGB潜力: {agb_increase_restoration_126:.4f} Pg DM")
print(f"高升温情境下恢复增加的AGB: {agb_increase_restoration_585:.4f} Pg DM")
print(f"高升温情境下恢复增加的AGB比低升温情境少: {difference:.4f} Pg DM, 相当于低升温情景潜力的 {percentage_less:.2f}%")
```
