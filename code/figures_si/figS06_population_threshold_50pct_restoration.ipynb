```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.mathtext import _mathtext as mathtext
mathtext.FontConstantsBase.sup1 = 0.45
import pickle

# Load data
with open('../../data/supplementary_figures/slr_res/DIVA_data/Coast_with_Mang.pkl', 'rb') as f:
    data = pickle.load(f)

# Set global font to Helvetica
plt.rcParams['font.family'] = 'Helvetica'
plt.rcParams['font.size'] = 16
# Set the mathtext font to Helvetica
plt.rcParams['mathtext.fontset'] = 'custom'
plt.rcParams['mathtext.rm'] = 'Helvetica'
plt.rcParams['mathtext.it'] = 'Helvetica:italic'
plt.rcParams['mathtext.bf'] = 'Helvetica:bold'


# --- MODIFICATION 1: Adjust figure layout ---
# Change from 3x2 to 2x2 grid and adjust figsize
fig, axs = plt.subplots(2, 2, figsize=(12, 12), gridspec_kw={'hspace': 0, 'wspace': 0}, constrained_layout=True)
fig.text(0, 0.5, 'Total AGB (Pg DM)', va='center', rotation='vertical', fontsize=18)

# --- MODIFICATION 2: Update threshold and add column indices ---
# Remove '5' and create a list of data column indices to use
threshold = ['0', '20']
slr_column_indices = [0, 2] # Column 0 is 'All', Column 2 is 'Pop20'

# --- MODIFICATION 3: Change loop range ---
for i in range(2):
    # --- MODIFICATION 4: Use the index list to select the correct data ---
    col_idx = slr_column_indices[i]
    
    # Select threshold and restoration strategy (50%)
    area_after_SLR_126_1 = data['area_after_SLR_126'][:, col_idx]
    area_after_SLR_585_1 = data['area_after_SLR_585'][:, col_idx]
    area_restoration_126_1 = data['area_restoration_126'][:, 0]
    area_restoration_585_1 = data['area_restoration_585'][:, 0]
    
    # Calculate total AGB
    total_agb_present = np.nansum(data['agb_present'] * data['area']) / 1e9
    total_agb_grow = np.nansum(data['agb_grow'] * data['area']) / (10 ** 9)
    total_agb_126 = np.nansum(data['agb_126'] * data['area']) / 1e9
    total_agb_585 = np.nansum(data['agb_585'] * data['area']) / 1e9
    total_agb_126_restoration = np.nansum(data['agb_126'] * (data['area'] + area_restoration_126_1)) / 1e9
    total_agb_585_restoration = np.nansum(data['agb_585'] * (data['area'] + area_restoration_585_1)) / 1e9
    total_agb_126_restoration_SLR = np.nansum(data['agb_126'] * (area_after_SLR_126_1 + area_restoration_126_1)) / 1e9
    total_agb_585_restoration_SLR = np.nansum(data['agb_585'] * (area_after_SLR_585_1 + area_restoration_585_1)) / 1e9

    '''Plot Low-Warming Scenario'''
    ax1 = axs[i, 0]

    xtips = np.arange(1, 7)
    ytips = [(total_agb_present + 1) / 2, total_agb_grow, total_agb_126, total_agb_126_restoration, total_agb_126_restoration_SLR, (total_agb_126_restoration_SLR + 1) / 2]
    labels = [f"{total_agb_present:.2f}",
              f"{total_agb_grow - total_agb_present:+.2f}",
              f"{total_agb_126 - total_agb_grow:+.2f}",
              f"{total_agb_126_restoration - total_agb_126:+.2f}",
              f"{total_agb_126_restoration_SLR-total_agb_126_restoration:+.2f}",
              f"{total_agb_126_restoration_SLR:.2f}"]
    bars = [
        ax1.bar(1, total_agb_present, color=(204 / 255, 204 / 255, 204 / 255), edgecolor='none'),
        ax1.bar(2, total_agb_grow, color=(118 / 255, 215 / 255, 196 / 255), edgecolor='none', label='Potential growth to maturity'),
        ax1.bar(2, total_agb_present, color='w', edgecolor='w'),
        ax1.bar(3, total_agb_126, color=(243 / 255, 182 / 255, 177 / 255), edgecolor='none', label='Climate and hydrology'),
        ax1.bar(3, total_agb_grow, color='w', edgecolor='w'),
        ax1.bar(4, total_agb_126_restoration, color=(235 / 255, 222 / 255, 247 / 255), edgecolor='none', label='Restoration'),
        ax1.bar(4, total_agb_126, color='w', edgecolor='w'),
        ax1.bar(5, total_agb_126_restoration, color=(174 / 255, 214 / 255, 241 / 255), edgecolor='none', label='Sea-level rise'),
        ax1.bar(5, total_agb_126_restoration_SLR, color='w', edgecolor='none'),
        ax1.bar(6, total_agb_126_restoration_SLR, color=(204 / 255, 204 / 255, 204 / 255), edgecolor='none')
    ]
    for j, (xtip, ytip, label) in enumerate(zip(xtips, ytips, labels)):
        label_value = float(label)
        if label_value > 0:
            ax1.text(xtip, ytip, label, ha='center', va='bottom')
        else:
            ax1.text(xtip, ytip - 0.01, label, ha='center', va='top')
            
    '''Plot High-Warming Scenario'''
    ax2 = axs[i, 1]
    xtips = np.arange(1, 7)
    ytips = [(total_agb_present + 1) / 2, total_agb_grow, total_agb_585, total_agb_585_restoration, total_agb_585_restoration_SLR, (total_agb_585_restoration_SLR + 1) / 2]
    labels = [f"{total_agb_present:.2f}",
              f"{total_agb_grow - total_agb_present:+.2f}",
              f"{total_agb_585 - total_agb_grow:+.2f}",
              f"{total_agb_585_restoration - total_agb_585:+.2f}",
              f"{total_agb_585_restoration_SLR-total_agb_585_restoration:+.2f}",
              f"{total_agb_585_restoration_SLR:.2f}"]
    bars = [
        ax2.bar(1, total_agb_present, color=(204 / 255, 204 / 255, 204 / 255), edgecolor='none'),
        ax2.bar(2, total_agb_grow, color=(118 / 255, 215 / 255, 196 / 255), edgecolor='none', label='Potential growth to mature'),
        ax2.bar(2, total_agb_present, color='w', edgecolor='w'),
        ax2.bar(3, total_agb_585, color=(243 / 255, 182 / 255, 177 / 255), edgecolor='none', label='Climate and hydrology'),
        ax2.bar(3, total_agb_grow, color='w', edgecolor='w'),
        ax2.bar(4, total_agb_585_restoration, color=(235 / 255, 222 / 255, 247 / 255), edgecolor='none', label='Restoration'),
        ax2.bar(4, total_agb_585, color='w', edgecolor='w'),
        ax2.bar(5, total_agb_585_restoration, color=(174 / 255, 214 / 255, 241 / 255), edgecolor='none', label='Sea-level rise'),
        ax2.bar(5, total_agb_585_restoration_SLR, color='w', edgecolor='none'),
        ax2.bar(6, total_agb_585_restoration_SLR, color=(204 / 255, 204 / 255, 204 / 255), edgecolor='none')
    ]
    for j, (xtip, ytip, label) in enumerate(zip(xtips, ytips, labels)):
        label_value = float(label)
        if label_value > 0:
            ax2.text(xtip, ytip, label, ha='center', va='bottom')
        else:
            ax2.text(xtip, ytip - 0.01, label, ha='center', va='top')

    # Subplot Config
    if i == 0:
        ax1.set_title(f"Low-Warming\n$\mathbf{{{chr(97+2*i)}}}$", loc='left')
        ax2.set_title(f"High-Warming\n$\mathbf{{{chr(98+2*i)}}}$", loc='left')
        
        handles, labels = ax1.get_legend_handles_labels()
        ax1.legend(handles, labels, loc='lower center', frameon=False, handlelength=0.8, fontsize=16, handletextpad=0.3)
    else:
        # The lettering (a, b, c, d) will be correct automatically
        ax1.set_title(f"$\mathbf{{{chr(97+2*i)}}}$", loc='left')
        ax2.set_title(f"$\mathbf{{{chr(98+2*i)}}}$", loc='left')

    # Set the axis
    for ax in [ax1, ax2]:
        ax.set_xlim([0, 7])
        ax.set_ylim([1.0, 2.5])
        ax.set_yticks(np.arange(1.0, 2.6, 0.5))
        ax.set_xticks([1, 6])
        ax.set_xticklabels(['Present-day', '2100'])
        ax.tick_params(axis='both')
        ax.spines['top'].set_visible(False)
        ax.spines['right'].set_visible(False)
        ax.spines['bottom'].set_visible(False)

    ax1.set_ylabel(' ', fontsize=18)

    ax2_right = ax2.twinx()
    # This now correctly refers to the updated 'threshold' list
    ax2_right.set_ylabel(f"Pop. density threshold = {threshold[i]} people km$^{{-2}}$", rotation=270, labelpad=20, color='k', fontsize=18)
    ax2_right.set_yticks([])
    ax2_right.spines['right'].set_visible(False)
    ax2_right.spines['top'].set_visible(False)
    ax2_right.spines['left'].set_visible(False)
    ax2_right.spines['bottom'].set_visible(False)

plt.savefig('../../figures/supplementary/figS06_population_threshold_50pct_restoration.png', dpi=400)
plt.show()
```
