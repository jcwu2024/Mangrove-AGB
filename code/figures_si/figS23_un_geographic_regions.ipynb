# Fig. S23. Geographic regions


```python
import geopandas as gpd
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

plt.rcParams['font.family'] = 'Helvetica'
plt.rcParams['font.size'] = 14
plt.rcParams['mathtext.fontset'] = 'custom'
plt.rcParams['mathtext.rm'] = 'Helvetica'
plt.rcParams['mathtext.it'] = 'Helvetica:italic'
plt.rcParams['mathtext.bf'] = 'Helvetica:bold'
```

```python
region_file = '../../data/supplementary_figures/region/world_regions.shp'
output_file = '../../figures/supplementary/figS23_un_geographic_regions.png'

regions = gpd.read_file(region_file)
unique_regions = [
    'Southeast Asia', 'South Asia', 'South America',
    'Western and Central Africa', 'North and Central America',
    'Oceania', 'Others'
]
custom_colors = ['#8DD3C7', '#FFFFB3', '#BEBADA', '#FB8072', '#80B1D3', '#FDB462', '#AFDA67']
region_colors = dict(zip(unique_regions, ListedColormap(custom_colors).colors))
```

```python
fig, ax = plt.subplots(
    figsize=(12, 3), dpi=800,
    subplot_kw={'projection': ccrs.PlateCarree()},
    constrained_layout=True
)
ax.set_extent([-180, 180, -40, 40])
ax.add_feature(cfeature.LAND, facecolor='lightgray')

for region, color in region_colors.items():
    regions[regions['RGN'] == region].plot(ax=ax, edgecolor='none', facecolor=color)

handles = [
    plt.Line2D([0], [0], marker='s', color='w', markerfacecolor=color, markersize=15)
    for color in region_colors.values()
]
ax.legend(handles, region_colors.keys(), loc='lower left', frameon=False)

fig.savefig(output_file, dpi=800)
plt.show()
```
