# Fig. S21. Present AGB uncertainty


```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
from matplotlib.mathtext import _mathtext as mathtext
import cartopy.crs as ccrs

mathtext.FontConstantsBase.sup1 = 0.45

plt.rcParams['font.family'] = 'Helvetica'
plt.rcParams['font.size'] = 16
plt.rcParams['mathtext.fontset'] = 'custom'
plt.rcParams['mathtext.rm'] = 'Helvetica'
plt.rcParams['mathtext.it'] = 'Helvetica:italic'
plt.rcParams['mathtext.bf'] = 'Helvetica:bold'
```

```python
agb_file = '../../data/supplementary_figures/model_results/AGB_upscale_2020_GB_halfdeg.csv'
uncertainty_file = '../../data/supplementary_figures/model_uncertainty/pixel_data_uncertainty_halfdeg.csv'
output_file = '../../figures/supplementary/figS21_present_agb_uncertainty.png'

map_df = pd.read_csv(agb_file)
map_df['MAD'] = pd.read_csv(uncertainty_file)['MAD']
map_df['MAD_p'] = map_df['MAD'] / map_df['AGB'] * 100
```

```python
fig = plt.figure(figsize=(12, 6), constrained_layout=True)
grid = fig.add_gridspec(2, 1)

ax1 = fig.add_subplot(grid[0], projection=ccrs.PlateCarree())
ax1.coastlines(resolution='110m', linewidth=0.75, color='black', zorder=0)
ax1.set_extent([-180, 180, -40, 40], crs=ccrs.PlateCarree())
ax1.set_title('a', loc='left', fontweight='bold')
scatter_mad = ax1.scatter(
    map_df['Lon'], map_df['Lat'], c=map_df['MAD'], cmap='RdYlBu_r', s=2,
    edgecolor='none', transform=ccrs.PlateCarree(), zorder=1
)
scatter_mad.set_clim(0, 100)
cax_top = inset_axes(
    ax1, width='100%', height='100%', loc='center left',
    bbox_to_anchor=(0.05, 0.15, 0.18, 0.08), bbox_transform=ax1.transAxes
)
cbar_top = plt.colorbar(scatter_mad, cax=cax_top, orientation='horizontal')
cbar_top.ax.set_title('Uncertainty (Quantified by mean \nabsolute deviation, $\pm$ Mg DM ha$^{-1}$)', fontsize=14, pad=10)
cbar_top.ax.tick_params(labelsize=14)

ax2 = fig.add_subplot(grid[1], projection=ccrs.PlateCarree())
ax2.coastlines(resolution='110m', linewidth=0.75, color='black', zorder=0)
ax2.set_extent([-180, 180, -40, 40], crs=ccrs.PlateCarree())
ax2.set_title('b', loc='left', fontweight='bold')
scatter_mad_p = ax2.scatter(
    map_df['Lon'], map_df['Lat'], c=map_df['MAD_p'], cmap='RdYlBu_r', s=2,
    edgecolor='none', transform=ccrs.PlateCarree(), zorder=1
)
scatter_mad_p.set_clim(0, 100)
cax_bottom = inset_axes(
    ax2, width='100%', height='100%', loc='center left',
    bbox_to_anchor=(0.05, 0.15, 0.18, 0.08), bbox_transform=ax2.transAxes
)
cbar_bottom = plt.colorbar(scatter_mad_p, cax=cax_bottom, orientation='horizontal')
cbar_bottom.ax.set_title('Uncertainty (Quantified by relative \naverage deviation, $\pm$ %)', fontsize=14, pad=10)
cbar_bottom.ax.tick_params(labelsize=14)

fig.savefig(output_file, dpi=600)
plt.show()
```
