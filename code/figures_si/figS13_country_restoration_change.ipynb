```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
country_agb_change_126 = pd.read_csv("../../data/supplementary_figures/slr_res/output/country_agb_change_126.csv")
country_agb_change_585 = pd.read_csv("../../data/supplementary_figures/slr_res/output/country_agb_change_585.csv")

# --- 数据重组与排序（与之前相同）---
df_126 = country_agb_change_126.rename(columns={'AGB_Change_RES': 'Low-warming'})
df_585 = country_agb_change_585.rename(columns={'AGB_Change_RES': 'High-warming'})
combined_data = pd.merge(df_126, df_585, on='Country')
combined_data['Decrease'] = combined_data['Low-warming'] - combined_data['High-warming']
sorted_data = combined_data.sort_values(by='Decrease', ascending=False)
plot_data = sorted_data.head(20)

# --- 绘图部分 (核心修改) ---

# Set global font to Helvetica
plt.rcParams['font.family'] = 'Helvetica'
plt.rcParams['font.size'] = 16
plt.rcParams['mathtext.rm'] = 'Helvetica'
plt.rcParams['mathtext.it'] = 'Helvetica:italic'
plt.rcParams['mathtext.bf'] = 'Helvetica:bold'

# 创建一个图和左侧坐标轴
fig, ax = plt.subplots(1, 1, figsize=(12, 6))

# --- 左侧Y轴: 绘制柱状图 ---
plot_data.plot(kind='bar', x='Country', y=['Low-warming', 'High-warming'], 
               ax=ax, color=['#b39ddb', '#80cbc4'])
ax.set_ylabel('AGB change from restoration (Tg DM)', fontsize=16)
ax.set_xlabel('')
ax.tick_params(axis='x', rotation=35, labelsize=14)
plt.xticks(ha='right')

# --- 右侧Y轴: 绘制差值散点图 ---
ax_right = ax.twinx() # 创建共享x轴的第二个y轴
x_coords = np.arange(len(plot_data))
point_color = '#c62828' # 定义一个颜色用于右轴和散点

ax_right.scatter(x=x_coords, y=plot_data['Decrease'], 
                 facecolor='none', edgecolor=point_color, s=80,
                 label='Difference', zorder=5, linewidth=1.5)

# 设置右轴的标签和颜色
ax_right.set_ylabel('Difference between low- and \nhigh-warming scenarios (Tg DM)', color=point_color, rotation = 270, labelpad = 40)
ax_right.tick_params(axis='y', labelcolor=point_color)
ax_right.spines['right'].set_color(point_color)


# --- 整合图例 ---
# 从两个坐标轴获取图例句柄(handles)和标签(labels)
h1, l1 = ax.get_legend_handles_labels()
h2, l2 = ax_right.get_legend_handles_labels()
# 合并并显示在一个图例中
ax.legend(h1 + h2, l1 + l2, frameon=False)
# 隐藏ax_right自身生成的图例，因为它已经被合并了
# ax_right.get_legend().remove()


# --- 全局优化 ---
# ax.set_title('Top 10 Countries by Largest Decrease in AGB Restoration Potential', loc='left', fontsize=20, pad=15)
ax.spines['top'].set_visible(False)
ax_right.spines['top'].set_visible(False) # 右轴的顶线也要隐藏

plt.tight_layout()
plt.savefig('../../figures/supplementary/figS13_country_restoration_change.png', dpi=300)
plt.show()
```
